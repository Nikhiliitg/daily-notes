```python
import torch as t
import torch.nn as nn

class PitballLoss(nn.Module):
    def __init__(self,tau=0.1):
        super(PitballLoss,self).__init__()
        self.tau = tau
        
    def forward(self,y_pred , y_true):
        error = y_pred - y_true
        return t.mean(
            self.tau * error , (1-self.tau)*error
            )
        
criterion = PitballLoss(tau=0.7)
```

```python
import numpy as np 
## Take it seriously its senior ML engineer thoughts
def stable_softmax(logits , target_idx):
    max_logit = np.max(logits)
    shifted_logits = logits - max_logit
    exp_logits = np.exp(shifted_logits)
    log_sum_exp = np.log(np.sum(np.exp(shifted_logits))) + max_logit
    return log_sum_exp - logits[target_idx]

logits = [1000,2,3,4]
target_idx = 0

print(stable_softmax(logits , target_idx))
```

```python
from math import sqrt
import torch as t
import torch.nn as nn
class LoRALayer:
    def __init__(self, in_dim , out_dim , rank , alpha):
        super(LoRALayer,self).__init__()
        self.rank = rank
        self.alpha = alpha
        self.scaling = alpha / sqrt(rank)
        
        self.lora_A = nn.Parameter(t.randn(rank , in_dim))
        self.lora_B = nn.parameter(t.zeros(out_dim , rank))
        
    def forward(self,x):
        update = (self.lora_B @ self.lora_A) @ x
        return self.base_layer(x) + self.scaling * update
    
```
